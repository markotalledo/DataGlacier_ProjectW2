## **Data Manipulation & Visualization**

### **Data Types and Customization**

```python
# We have to change the data type to string
data['Customer ID'] = data['Customer ID'].astype('string')
data['Transaction ID'] = data['Transaction ID'].astype('string')
```

```python
# Income got renamed
data.rename(columns={'Income (USD/Month)': 'Income'}, inplace=True)
```

```python
# Transforming to date format
data['Date of Travel'] = data['Date of Travel'].apply(
    lambda s: xlrd.xldate.xldate_as_datetime(s, 0))
```

```python
# Transforming strings to floats
data['Population'] = data['Population'].str.replace(',', '').astype(float)
data['Users'] = data['Users'].str.replace(',', '').astype(float)
```

### **New Variables and KPIS**

```python
# Profit variable - We need to see what amount of money is gained per service
data['Profit'] = data['Price Charged'] - data['Cost of Trip']
data['Profit']
```

```python
# Unit Cost per KM Travelled
data['cost_per_km'] = data['Cost of Trip'] / data['KM Travelled']
data['cost_per_km']
```

```python
# Profit that is gained per KM Travelled
data['profit_per_km'] = data['Profit'] / data['KM Travelled']
data['profit_per_km']
```

### **Outliers Treatment**

```python
# There are many outliers in the 'Price Charged' Variable, it is better to
# drop them in order to clean our data. Interquartile Range method is going to
# be used

Q1 = data['Price Charged'].quantile(0.25)
Q3 = data['Price Charged'].quantile(0.75)

IQR = Q3 - Q1

data = data[(data['Price Charged'] >= (Q1 - 1.5*IQR)) &
            (data['Price Charged'] <= (Q3 + 1.5*IQR))]
```

```python
# REMOVING OUTLIER
Q1 = data['profit_per_km'].quantile(0.25)
Q3 = data['profit_per_km'].quantile(0.75)

IQR = Q3 - Q1
IQR

data = data[(data['profit_per_km'] >= (Q1 - 1.5*IQR)) &
            (data['profit_per_km'] <= (Q3 + 1.5*IQR))]


# %%

# REMOVING OUTLIERS
Q1 = data['Profit'].quantile(0.25)
Q3 = data['Profit'].quantile(0.75)

IQR = Q3 - Q1
IQR

data = data[(data['Profit'] >= (Q1 - 1.5*IQR)) &
            (data['Profit'] <= (Q3 + 1.5*IQR))]
```

```python
# REMOVING OUTLIERS PER CITY
for c in data['City'].unique():
    Q1 = data.loc[data['City'] == c, ['Profit']].quantile(0.25)
    Q3 = data.loc[data['City'] == c, ['Profit']].quantile(0.75)
    IQR = Q3 - Q1
    LL = float(Q1 - 1.5*IQR)
    RL = float(Q3 + 1.5*IQR)
    cond1 = (data['City'] == c) & (data['Profit'] >= LL)
    cond2 = (data['City'] == c) & (data['Profit'] <= RL)
    if c == data['City'].unique()[0]:
        outdata = data.loc[cond1 & cond2, :]
    else:
        dummy = data.loc[cond1 & cond2, :]
        outdata = pd.concat([outdata, dummy])
    print(outdata.shape)
```

```python
# REMOVING OUTLIERS PER CITY

for c in outdata['City'].unique():
    Q1 = outdata.loc[outdata['City'] == c, ['profit_per_km']].quantile(0.25)
    Q3 = outdata.loc[outdata['City'] == c, ['profit_per_km']].quantile(0.75)
    IQR = Q3 - Q1
    LL = float(Q1 - 1.5*IQR)
    RL = float(Q3 + 1.5*IQR)
    cond1 = (outdata['City'] == c) & (outdata['profit_per_km'] >= LL)
    cond2 = (outdata['City'] == c) & (outdata['profit_per_km'] <= RL)
    if c == outdata['City'].unique()[0]:
        data_out = outdata.loc[cond1 & cond2, :]
    else:
        dummy = outdata.loc[cond1 & cond2, :]
        data_out = pd.concat([data_out, dummy])
    print(data_out.shape)
```

```python
# %%
cost_medians = data_out.groupby('City')['cost_per_km'].median()


# %%
# STANDARIZING COST PER KM FOR EACH CITY

for cit in data_out['City'].unique():
    data_out.loc[data_out['City'] == cit, ['cost_per_km']] = cost_medians[cit]



# We have decided to choose standard fees because each city has their own cost structure: Taxes, Fees, etc.
# That's why we simplified those fees with the median of its values
```

### **Time Maniulation**

```python
# Now, We create a new dataframe and set the date column as the index
timedata = data_out.copy()
timedata = data_out.set_index('Date of Travel')
timedata.head()

# The dates that we have are from 2016 toi 2018 period
```

```python
# OUR TARGET VARIABLE IS THE PROFIT VARIABLE, SO LET'S WATCH HOW THE MEAN CHANGED THROUGH TIME

df_month = timedata.resample("M").mean()
fig, ax = plt.subplots(figsize=(10, 6))
ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m'))
ax.bar(df_month['2016':].index, df_month.loc['2016':,
       "Profit"], width=25, align='center')

# As we said before, the lowest point starts on september and then the profit arises until January and May
```

```python
# %%
# Then we need to extract the Year and Month from the dates 
# In order to make a deeeper analysis of the differences from month to month
timedata['Month'] = pd.DatetimeIndex(timedata.index).month
timedata['Year'] = pd.DatetimeIndex(timedata.index).year
```

```python

# %%
# First, We observed our numeric variables and how they are distributed through months,
fig, axes = plt.subplots(4, 1, figsize=(10, 16), sharex=True)
for name, ax in zip(['Profit', 'profit_per_km', 'KM Travelled', 'Cost of Trip'], axes):
    sns.boxplot(data=timedata, x='Month', y=name, ax=ax)
    ax.set_ylabel("")
    ax.set_title(name)
    if ax != axes[-1]:
        ax.set_xlabel('')


# KM Travelled and Cost of Trip have the same variation through months, analyze that part won't help too much
# We need to focus in the changes of the Profit and profit_per_km variables because the variations that they present
```

```python


# %%
df_month['Profit'].plot(figsize=(8, 6))

# Again, July are the lowest profit month because the holidays that americans celebrate
# And, January and December are the greatest profits because Christmas and New year's eve.
```

```python

# %%
# To make a deeper analysis, we need to focus in the KM Travelled between months
timedata.groupby(['Year', 'Month'])[
    'KM Travelled'].mean().plot(figsize=(8, 6))


# The KM Travelled get their highest points on July, that tell us that exists sales and offers that people uses in order to
# take rides and increase the demand for both companies
```

```python
# %%

# Then, we check the difference between month to month of the profit 
# We confirm the seasonality, the biggest falls are situated on July
df_month.loc[:, 'pct_change'] = df_month.Profit.pct_change()*100
fig, ax = plt.subplots()
df_month['pct_change'].plot(kind='bar', color='coral', ax=ax)
ax.xaxis.set_major_locator(mdates.MonthLocator())
plt.xticks(rotation=45)
ax.legend()
```
